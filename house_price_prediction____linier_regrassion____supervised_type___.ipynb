```python
cdf = df[[   'price' ,	'bedrooms',	'bathrooms',	'sqft_living',	'sqft_lot' ,	'floors',	'condition' ,	'sqft_above' ,	'sqft_basement'	,'yr_built' ,	'yr_renovated']]
cdf.head(800)
```

```python
plt.scatter(cdf.sqft_lot, cdf.price,  color='green')
plt.xlabel("sqft_lot")
plt.ylabel("price")
plt.show()
```

#   #test trian split
## Divide the data set in to test_data_set & train_data_set

```python
msk = np.random.rand(len(df)) < 0.8
train = cdf[msk]
test = cdf[~msk]
```

##plot graph between  train data set (x_train , y_train) by scatter_ method

|||DATA VISULZATION|||

```python
plt.scatter(train.sqft_lot, train.price,  color='green')
plt.xlabel("sqft_lot")
plt.ylabel("price")
plt.show()
```

#from  sklearn libary import the liner_model

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
```

create array

```python

train_x = np.asanyarray(train[['sqft_lot']])
train_y = np.asanyarray(train[["price"]])

#fit / impliment test train data in obj (regr)

regr.fit (train_x, train_y)

#  print The coefficients
print ('Coefficients: ', regr.coef_)
#  print The Intercept
print ('Intercept: ',regr.intercept_)
```

# #plot the graph between train sqft_lot ,&train price  
## |||data visuzilation |||

```python
plt.scatter(train.sqft_lot, train.price,  color='blue')
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], '-r')
plt.xlabel("sqft_lot")
plt.ylabel("price")
```

```python
#  answer

#find a house price for 8000sqft_lot land


y_pred=regr.predict([[8000]])
```
